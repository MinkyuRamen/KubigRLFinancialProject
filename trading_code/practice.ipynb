```python
import pandas as pd
import numpy as np
import yfinance as yf

import matplotlib.pyplot as plt
import seaborn as sns
```

# data loading

```python
ticker = ['AAPL','MSFT','AMZN','GOOGL','TSLA']
# ticker = ['AAPL','MSFT','AMZN','GOOGL','NVDA','META','TSLA']
stocks = yf.download(ticker, start = "2015-01-01", end = "2023-04-30")
```

```python
stocks.head()
```

```python
stocks.to_csv("stocks_top5.csv")
```

```python
stocks = pd.read_csv("stocks_top5.csv", header = [0, 1], index_col = [0], parse_dates = [0])
stocks.head()
```

```python
sort_stocks = stocks.swaplevel(axis = 1).sort_index(axis = 1)
sort_stocks.head()
```

```python
closevol = stocks.loc[:, ["Adj Close","Volume"]].copy()
closevol.head()
```

```python
closevol['Adj Close'].plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

```python
closevol['Volume'].plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

```python
close = closevol['Adj Close']
norm = close.div(close.iloc[0]).mul(100)
norm.head()
```

```python
norm.plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```
