## Streaming high-frequency real-time Data

```python
import pandas as pd
import fxcmpy
```

```python
api = fxcmpy.fxcmpy(config_file= "fxcm.cfg")
```

```python
api.subscribe_market_data("BTC/USD")
```

```python
api.get_subscribed_symbols()
```

```python
api.get_last_price("BTC/USD")
```

```python
api.get_prices("BTC/USD")
```

```python
import time
```

```python
while True:
    time.sleep(1)
    print(api.get_last_price("EUR/USD").name, api.get_last_price("EUR/USD").Bid,
          api.get_last_price("EUR/USD").Ask, sep = "|")
```

```python
api.unsubscribe_market_data("EUR/USD")
```

```python
def print_data(data, dataframe):
    print('%3d | %s | %s, %s, %s'
          % (len(dataframe), data['Symbol'],
             pd.to_datetime(int(data['Updated']), unit='ms'),
             data['Rates'][0], data['Rates'][1]))
```

```python
api.subscribe_market_data("EUR/USD", (print_data,))
```

```python
api.unsubscribe_market_data("EUR/USD")
```

```python
api.close()
```

## Creating Orders and Executing Trades

```python
import pandas as pd
import fxcmpy
```

```python
api = fxcmpy.fxcmpy(config_file= "fxcm.cfg")
```

```python
api.get_open_positions() # open position 가져온다.
```

```python
api.create_market_buy_order("BTC/USD", 100) # BTC/USD 100unit 만큼 long position
```

```python
api.get_open_positions() # long position 뜬다.
```

```python
col = ["tradeId", "amountK", "currency", "grossPL", "isBuy"]
```

```python
api.get_open_positions()[col] # open positions 정보 불러온다.
```

```python
api.create_market_sell_order("BTC/USD", 100) # BTC/USD 100unit 만큼 short position
```

```python
api.get_open_positions()[col] # long & short 둘다 가질 수 있는 헷지 계좌이다.
```

```python
api.close_all_for_symbol("BTC/USD") # 모든 position 청산
```

```python
api.get_closed_positions()[col] # 청산한 position 가져온다.
```

```python
api.get_closed_positions_summary()[col] # total grossPL 보여준다.
```
